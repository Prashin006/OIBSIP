### Prepare Our Data for Modelling
##### This involves splitting the data back out into plain NumPy arrays

```python
X_train = df_train.drop(columns=['Species','Species_name']).values
# X_train,type(X_train),X_train.shape
y_train = df_train['Species'].values
# y_train.shape
```

### Baseline Modelling
##### It is the simplest model we can think of. In this case it can be random guessing of the species, or guessing a single species for every data point (33.33% accuracy), or based on the linear separability as given in the dataset (that we can correctly separate one class from other 2 but not all the 3 classes. So our model must beat atleast 33.33% accuracy.

### Simple Manual Model
##### In the pairplots above we can see that if petal length is less than 2.4 / 2.5 then it is setosa & if it is less than 4.75 but greater than 2.5 then it is versicolor else it is virginica

```python
def single_feature_prediction(petal_length):
    '''Predicts the Iris Species given the petal length. Similar to Decision Tree'''
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.75:
        return 1
    else:
        return 2
```

```python
# df_train['Petal Length'].values
X_train[:,2]
```

```python
manual_y_pred = np.array([single_feature_prediction(val) for val in X_train[:,2]])
manual_y_pred
```

```python
y_train
```

```python
manual_y_pred == y_train
```

```python
manual_model_acc = np.mean(manual_y_pred == y_train)*100    # To get accuracy, np.mean() treats True as 1 & false as 0
manual_model_acc.round(3)
```

```python
print(f"Manual Model (Decision Tree) accuracy: {manual_model_acc:.3f}%")
```

```python
def plot_incorrect_predictions(df_predictions,x_axis_feature,y_axis_feature):
    fig,axs = plt.subplots(2,2,figsize=(10,10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue='prediction_label',data=df_predictions,ax=axs[0])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue='Species_name',data=df_predictions,ax=axs[1])
    sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue='correct_prediction',data=df_predictions,ax=axs[2])
    # sns.scatterplot(x=x_axis_feature,y=y_axis_feature,hue='prediction',data=df_predictions,ax=axs[3])
    axs[3].set_visible(False)
    plt.show()
```

```python
fig,axs = plt.subplots(2,2,figsize=(10,10))
axs,axs.shape
```

```python
plot_incorrect_predictions(df_predictions=df_predictions,x_axis_feature='Petal Length',y_axis_feature='Petal Width')
```
